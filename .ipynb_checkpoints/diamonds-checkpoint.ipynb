## Load libraries and data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.ion()
import seaborn as sns
# import os

diamonds = sns.load_dataset("diamonds")
```

## Prepare data for descriptive analysis

```python
from pandas.api.types import CategoricalDtype

# Function that prepares a pandas df line by line
#  -> handling of categories looks too verbose but allows (A) for safe preparation of a single line
#     and (B) for the right ordering of levels.
def transform(df):
    df = df.assign(log_price=np.log(df.price),
                   log_carat=np.log(df.carat),
                   depth=np.clip(df.depth, 50, 75),
                   table=np.clip(df.table, 50, 70),
                   color=pd.Categorical(df.color, categories=['D', 'E', 'F', 'G', 'H', 'I', 'J']),
                   cut=pd.Categorical(df.cut, categories=['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']),
                   clarity=pd.Categorical(df.clarity, categories=['IF', 'VVS1', 'VVS2', 'VS1', 'VS2', 'SI1', 'SI2', 'I1']))
    return df

df = transform(diamonds)

# Important variable groups
response = "log_price"
num_vars = ["log_carat", "depth", "table"]

ordinal_vars = ["color", "cut", "clarity"]

features = num_vars + ordinal_vars
df.head()
```

## Prepare data for modeling

```python
# Safe preprocessing with pands
def ordinalCoder(df, v=df.columns):
    df = df.copy()
    df[v] = df[v].apply(lambda x: x.cat.codes)
    return df
    
# Response & features
y = df[response]
X_display = df[features]

# Integer coding
X = ordinalCoder(X_display, ordinal_vars)

print("Data structure for model:\n\n", X.head())
print("Data structure for interpretation:\n\n", X_display.head())
```

```python
# For one obs, it is working?
step1 = transform(diamonds.head(1))
print(step1)
print(ordinalCoder(step1, ordinal_vars))
```

## Modeling

### Data split

```python
# Data split
from sklearn.model_selection import train_test_split
import lightgbm as lgb

# create a train/test split
X_display_train, X_display_test, X_train, X_test, y_train, y_test = train_test_split(
    X_display, X, y, test_size=0.2, random_state=7, shuffle=True)

dtrain = lgb.Dataset(X_train, 
                     label=y_train, 
                     free_raw_data=False)
```

### Fit model

```python
# Reasonable params
params = {
        "learning_rate": 0.02,
        "num_leaves": 15,
        'min_child_samples': 10,
        'reg_alpha': 1,
        'reg_lambda': 0,
        'min_split_gain': 0.0000,
        'subsample': 0.9,
        'bagging_freq': 1,
        'colsample_bytree': 1,
        "verbose": 2,
        "objective": "regression",
        "metric": "rmse",
        "nthread": 8
}

# Fit the model
model = lgb.train(params, dtrain, 4535)
```

### Performance

RMSE on both training and test data.

```python
# Performance (ROC Plot)
from sklearn.metrics import mean_squared_error as mse
pred_train = model.predict(X_train)
pred_test = model.predict(X_test)

print("Training RMSE:", np.sqrt(mse(y_train, pred_train)))
print("Test RMSE:", np.sqrt(mse(y_test, pred_test)))
```

## Explain model

```python
import shap

# print the JS visualization code to the notebook
shap.initjs()

# Number of SHAP decompositions
N_SHAP = 2000 

# Initialize tree explainer
explainer = shap.TreeExplainer(model)

# Calculate N_SHAP decompositions
shap_values = explainer.shap_values(X_train[:N_SHAP])
```

### Variable importance

```python
shap.summary_plot(shap_values, X_train[:N_SHAP], plot_type="bar")
```

### Effects plots

```python
sorted_cols = X_train.columns[np.argsort(-np.abs(shap_values).mean(0))]

for name in sorted_cols:
    shap.dependence_plot(name, shap_values, X_train[:N_SHAP], 
                         display_features=X_display_train[:N_SHAP], 
                         show=False, alpha=0.1, dot_size=4, x_jitter=0.1, 
                         xmin="percentile(1)", xmax="percentile(99)")
    plt.title(f"Dependence plot - {name}", fontdict={'fontsize': 15})
    plt.show()
```
